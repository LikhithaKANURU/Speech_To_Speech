```python
#TEXT TO SPEECH
import pyttsx3

class TextToSpeech:
    def __init__(self):
        self.engine = pyttsx3.init()
        self.engine.setProperty('rate', 120)
        voices = self.engine.getProperty('voices')
        self.engine.setProperty('voice', voices[1].id)  # Use the correct voice property

    def speak(self, text):
        self.engine.say(text)
        self.engine.runAndWait()
        
tts = TextToSpeech()
while True:
    text = input("Enter text to convert to speech (type 'exit' to quit): ")
    if text.lower() == 'exit':
        break
    tts.speak(text)
```

```python
import pyttsx3 as p
import speech_recognition as sr
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

# Define the Infow class
class Infow:
    def __init__(self):
        # Set up the ChromeDriver service and options
        chrome_options = Options()
        service = Service(ChromeDriverManager().install())

        # Initialize the WebDriver
        self.driver = webdriver.Chrome(service=service, options=chrome_options)

    def get_info(self, query):
        self.query = query
        # Load the Wikipedia homepage
        self.driver.get("https://www.wikipedia.org")
        # Find the search input element and enter the query
        search = self.driver.find_element(By.XPATH, '//*[@id="searchInput"]')
        search.click()
        search.send_keys(query)
        # Find the search button and click it
        enter = self.driver.find_element(By.XPATH, '//*[@id="search-form"]/fieldset/button')
        enter.click()

# Initialize the text-to-speech engine
engine = p.init()
rate = engine.getProperty('rate')
engine.setProperty('rate', 120)
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)  # Use the correct voice property

def speak(text):
    engine.say(text)
    engine.runAndWait()

# Initialize the recognizer
r = sr.Recognizer()

def listen_for_command():
    with sr.Microphone() as source:
        r.energy_threshold = 10000
        r.adjust_for_ambient_noise(source, duration=1.2)
        print("Listening...")
        try:
            audio = r.listen(source)
            text = r.recognize_google(audio)
            return text
        except sr.UnknownValueError:
            print("Sorry, I did not understand that.")
            speak("Sorry, I did not understand that.")
        except sr.RequestError:
            print("Sorry, there was an error with the speech recognition service.")
            speak("Sorry, there was an error with the speech recognition service.")
        return ""

speak("Hello Madam, I am your voice assistant. How are you?")
text = listen_for_command()

if "what" in text and "about" in text and "you" in text:
    speak("I am also having a good day Madam")

speak("What can I do for you Madam?")
text2 = listen_for_command()

if "information" in text2:
    speak("You need information related to which topic?")
    infor = listen_for_command()
    if infor:
        speak(f"Searching in Wikipedia for {infor}")
        assist = Infow()
        assist.get_info(infor)
```
